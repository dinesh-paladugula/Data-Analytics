```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Compute 'Total Spent' column (if not already present in the dataset)
# df['Total Spent'] = df['Quantity'] * df['Unit Price']

# Select numerical columns for correlation analysis
numerical_columns = df.select_dtypes(include=[np.number])  # Get only numeric columns
correlation_matrix = numerical_columns.corr()  # Compute the correlation matrix

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Analyze total spending by gender
total_spent_by_gender = df.groupby('Gender')['Total Amount'].mean()
print("Mean Total Spent by Gender:")
print(total_spent_by_gender)

# Violin plot to visualize the distribution of spending by gender
sns.violinplot(x='Gender', y='Total Amount', data=df, palette='pastel', legend=False)
plt.title('Total Spent Distribution by Gender')
plt.ylabel('Total Spent')
plt.xlabel('Gender')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataset has columns 'Product Category' and 'Unit Price'
# Group by Product Category and calculate the mean Unit Price
price_per_category = df.groupby('Product Category')['Price per Unit'].mean().reset_index()

# Pivot data for heatmap, set index and value (if more than one category dimension exists)
heatmap_data = price_per_category.pivot_table(index='Product Category', values='Price per Unit')

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap: Price Per Unit Across Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Average Unit Price')
plt.show()
```
